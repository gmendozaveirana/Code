```python
# Import
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import math
import ptf
import MCMCgraphs
import random
import seaborn as sns
e_0 = 8.8541878128e-12
pd.set_option('display.max_columns', None)
```

```python
'''####################### Lists for synthetic evaluation of pedophysical models ######################'''
step =     500000
vmc_ =     np.arange(0.04,    .5  , (.5    - 0.04)  /step)
bd_ =      np.arange(1.1 ,    1.82 , (1.82  - 1.1) /step)
pd_ =      np.arange(2.46,    2.86, (2.86  - 2.46)  /step)
cc_ =      np.arange(2   ,    90  , (90    - 2)     /step)
conds =    np.arange(.001,    0.5 , (0.5   - .001)  /step)
silt_ =    np.arange(1   ,    100 , (100   - 1)     /step)
sc_ =      np.arange(1   ,    100 , (100   - 1)     /step)
org_ =     np.arange(0.2   ,  2   , (2     - 0.2)   /step)
oc_ =      np.arange(1   ,    15  , (15    - 1)     /step)
lois =     np.arange(2   ,    300 , (300   - 2)     /step)
cec_ =     np.arange(1.6,     40,   (40    - 1.6)   /step)
L_ =       np.arange(0.1 ,    0.9 , (0.9   - 0.1)   /step)
freqgpr_ = np.arange(30e6,    1.3e9,(1.3e9 - 30e6)  /step)
freqcon_ = np.arange(1 ,      30e6, (30e6  - 1    ) /step)
freq_ =    np.arange(1,       300e6,(300e6 - 1)     /step)
t_      =  np.arange(1      , 50  , (50    - 1)     /step)
wp_ =      np.arange(78  ,    82 ,  (82    - 78)    /step)
ap_ =      np.arange(1  ,    1.4 ,  (0.4)    /step)

cm_ =      np.arange(0.0001, 0.005, (0.005 - 0.0001)/step) 
ps_ =      np.arange(3   ,    4 ,   (4     - 3)     /step)
offset_ =  np.arange(3   ,    6 ,   (6     - 3)     /step)

rsw_ =     np.arange(1   ,    50  , (50    - 1)     /step)
depths =   np.arange(1      , 100 , (100   - 1)     /step)
m_ =       np.arange(0.8    ,   1.8   , (1)     /step)
n_ =       np.arange(1.5    , 2.5 , (2.5   - 1.5)   /step)
por_ =     np.arange(0.04,    .50 , (0.5   - 0.04)  /step)

ri_ =      np.arange(1350,    2650, (2650  - 1350)  /step)
wi_ =      np.arange(0.025,   0.087,(0.087 - 0.025) /step)
Lwc_ =     np.arange(0.001 ,  0.3 ,(0.3  - 0.001) /step)

L_graph_wund =  np.ones((len(vmc_)))*0.18
L_graph_endr =  np.ones((len(vmc_)))*0.31
L_graph_fengs = np.ones((len(vmc_)))*0.3


""" ##############################  Random shuffle of the above lists  ################################"""

rvmc_    = random.sample(list(vmc_),  len(vmc_))
rbd_     = random.sample(list(bd_ ),  len(bd_ ))
rpd_     = random.sample(list(pd_ ),  len(pd_ ))
rcc_     = random.sample(list(cc_ ),  len(cc_ ))
rL_      = random.sample(list(L_  ),  len(L_  ))
rcm_     = random.sample(list(cm_ ),  len(cm_ ))
rm_      = random.sample(list(m_  ),  len(m_  ))
rn_      = random.sample(list(n_  ),  len(n_  ))
rrsw_    = random.sample(list(rsw_),  len(rsw_))
rps_     = random.sample(list(ps_ ),  len(ps_ ))
rorg_    = random.sample(list(org_),  len(org_))
rwp_     = random.sample(list(wp_ ),  len(wp_ ))
roffset_ = random.sample(list(offset_),  len(offset_))

rLwc_ = random.sample(list(Lwc_),  len(Lwc_))
rri_  = random.sample(list(ri_ ),  len(ri_ ))
rwi_  = random.sample(list(wi_ ),  len(wi_ ))
print(type(vmc_))
print(type(list(vmc_)))
print(type(rvmc_))
```

```python
Info = pd.DataFrame(columns = [ " Conductivity Models ", "CombArchie", "LR1",   "WundC", "Glover101", 
                                " Permittivity Models ", "CRIM",       "Sen81", "Roth1", "RothIOP1", "fengsen1985",
                                                         "LindeP",     "ModLindeP"                       ],
                      index = ["Stand_dev", "Mean"])
Log_scale = "no"
```
