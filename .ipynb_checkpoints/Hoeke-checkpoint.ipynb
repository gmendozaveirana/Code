```python
"""     Introducing corrections    """

df["BulkCondCorrec"] =     (ptf.logsdon(50000000, df.RawRealDielectric[:],   df.RawImDielectric[:], np))
df["RelaxationC"] =        df.BulkECTCSm[:]   - df.BulkCondCorrec[:]
df["RelaxationP"] =        df.RawImDielectric[:]   - df.BulkCondCorrec[:]/(2*pi*fc*e_0)
df['WatpermT'] =           ptf.jones05(df.SoilTemperatureC)
df["WaterCondCorrec"] =    (df.WatpermT*df.BulkCondCorrec[:])/(df.RawRealDielectric[:] - df.Offset)
df["AparentP"] =           (df.RawRealDielectric[:]/2)*(1+(1+(df.RawImDielectric[:]/df.RawRealDielectric[:])**2)**0.5)
```

```python
ps = 3.67

dt['WatpermT'] =           ptf.Malmberg_Maryott56(dt.field_temp)
dt["partdens"] = ptf.schjonnpd(dt.Clay, dt.Humus)

dt["realperm_pred_CRIM"] = ptf.CRIM(dt.field_moisture/100, dt.Bulk_density, dt.partdens, 1.4, ps, dt.WatpermT)/e_0
dt["realperm_pred_roth_IOP"] = ptf.roth1990_IOP(dt.field_moisture/100, dt.Bulk_density, dt.partdens, 1.4, ps, dt.WatpermT, dt.CEC_meq100g)/e_0

dt["realperm_pred_lindep"] = ptf.lindeperm(dt.field_moisture/100, dt.Bulk_density, dt.partdens, 1.4, ps, dt.WatpermT, 1.5, 2)/e_0
dt["realperm_pred_lindep_IOP"] = ptf.lindep_IOP(dt.field_moisture/100, dt.Bulk_density, dt.partdens, 1.4, ps, dt.WatpermT, dt.CEC_meq100g)/e_0

dt["realperm_pred_sen"] = ptf.sen1981(dt.field_moisture/100, dt.Bulk_density, dt.partdens, 0.3, 1.4, ps, dt.WatpermT)/e_0
dt["realperm_pred_sen_IOP"] = ptf.sen81_IOP(dt.field_moisture/100, dt.Bulk_density, dt.partdens, 1.4, ps, dt.WatpermT, dt.CEC_meq100g)/e_0

dt["wund13_IOP_pred"] = ptf.wund13_IOP(dt.field_moisture/100,  5, 0.005, dt.WatpermT, dt.CEC_meq100g)/e_0
dt["wunderlich2013perm_pred"] = ptf.wunderlich2013perm(dt.field_moisture/100, 5, 0.05, 0.01, dt.WatpermT)/e_0


dt["perm_pred_Hydra"] = ptf.Hydraprobe(dt.field_moisture/100)/e_0

dt["endresredman_IOP"]   =  ptf.endresredman_IOP(dt.field_moisture/100, dt.Bulk_density, dt.partdens, 1.4, ps, dt.WatpermT, dt.CEC_meq100g)/e_0
dt["endresredman"]   =  ptf.endresredman1996(dt.field_moisture/100, dt.Bulk_density, dt.partdens, 0.3, 1.4, ps, dt.WatpermT)/e_0

dt["FengSen_IOP"]   =  ptf.fengsen_IOP(dt.field_moisture/100, dt.Bulk_density, dt.partdens, 1.4, ps, dt.WatpermT, dt.CEC_meq100g)/e_0
dt["FengSen"]   =  ptf.fengsen1985(dt.field_moisture/100, dt.Bulk_density, dt.partdens, 0.3, 1.4, ps, dt.WatpermT)/e_0

dt.describe()


Fits.at["CRIM", "R2"] = r2_score(dt.realperm_pred_CRIM, dt.field_realperm)
Fits.at["Roth_IOP", "R2"] = r2_score(dt.realperm_pred_roth_IOP, dt.field_realperm)

Fits.at["perm_pred_Hydra", "R2"] = r2_score(dt.perm_pred_Hydra, dt.field_realperm)
Fits.at["Lindep_IOP", "R2"] = r2_score(dt.realperm_pred_lindep_IOP, dt.field_realperm)
Fits.at["Lindep", "R2"] = r2_score(dt.realperm_pred_lindep, dt.field_realperm)

Fits.at["Sen81", "R2"] = r2_score(dt.realperm_pred_sen, dt.field_realperm)
Fits.at["Sen81_IOP", "R2"] = r2_score(dt.realperm_pred_sen_IOP, dt.field_realperm)

Fits.at["Wund", "R2"] = r2_score(dt.wunderlich2013perm_pred, dt.field_realperm)
Fits.at["Wund_IOP", "R2"] = r2_score(dt.wund13_IOP_pred, dt.field_realperm)

Fits.at["endresredman_IOP", "R2"] = r2_score(dt.endresredman_IOP, dt.field_realperm)
Fits.at["endresredman", "R2"] = r2_score(dt.endresredman, dt.field_realperm)

Fits.at["FengSen_IOP", "R2"] = r2_score(dt.FengSen_IOP, dt.field_realperm)
Fits.at["FengSen", "R2"] = r2_score(dt.FengSen, dt.field_realperm)

####
Fits.at["CRIM", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.realperm_pred_CRIM))
Fits.at["Roth_IOP", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.realperm_pred_roth_IOP))


Fits.at["perm_pred_Hydra", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.perm_pred_Hydra))

Fits.at["Lindep_IOP", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.realperm_pred_lindep_IOP))
Fits.at["Lindep", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.realperm_pred_lindep))


Fits.at["Sen81", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.realperm_pred_sen))
Fits.at["Sen81_IOP", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.realperm_pred_sen_IOP))

Fits.at["Wund", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.wunderlich2013perm_pred))
Fits.at["Wund_IOP", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.wund13_IOP_pred))

Fits.at["endresredman", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.endresredman))
Fits.at["endresredman_IOP", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.endresredman_IOP))

Fits.at["FengSen", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.FengSen))
Fits.at["FengSen_IOP", "RMSE"] = math.sqrt(mean_squared_error(dt.field_realperm, dt.FengSen_IOP))

Fits
```
