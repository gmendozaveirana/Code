```python
"""
################################## PERMITTIVITY PEDOPHYSICAL MODELS EVALUATION ##################################
"""
""" 
Defining a few fix parameters 
"""
ap =       1.3                                              # Air permittivity phase
p_inf =    3.5                                              # Permittivity relative of the soil at infinity frecuency

"""
Defining default synthetic soil physical parameters

"""
vmc_ =           np.arange(0.0025, .5 , (.5 - 0.0025)/100)  # Volumetric water content
alpha_ =         np.arange(0.01,   1.5, (1.5-0.01 )  /100)  # All posible alpha values (geometric parameter)
m_ =             np.arange(0.5,   2,    (2 - 0.5)    /100)  # All posible cementation factor (m) values (geometric parameter)
bwp_ =           np.arange(-5000, 2000, 7000         /100)  # All posible bound water values 

L_ =             np.arange(-0.2,  0.8,  (1)          /50 )  # All posible depolarization factor (L) values (geometric parameter)
Lw_ =            np.arange(-0.1,  0.2,  (0.3)        /50 )  # All posible depolarization factor (Lw) values (geometric parameter)
crim_pred_, Roth90_mv_ = [], []

samples = ds['Sample_name'].to_list()[:]                   # List of sample's names

s = 0
for sample in samples:                                      # This iterates over all the ten samples       
    
    ############################## Reading some parameters for each soil sample ######################
    
    print("Sample:", sample)
    sc =      ds.at[s, "Sand_content"]               # SAND content (63 - 2000 * e-6 meters) (%)
    silt =    ds.at[s, "Silt_content"]               # SILT content (%)
    cc  =     ds.at[s, "Clay_content"]               # CLAY content (0  - 2    * e-6 meters) %[m3/m3]
    bd =      ds.at[s, "Bulk_density_reached"]       # Bulk density reached during the laboratory sample preparation the same as measured in field conditions.
    ps =      ds.at[s, "Solid_phase_permittivity"]   # Solid phase permittivity calculated using crim model, after experiment.
    org =     ds.at[s, "Humus"]                      # Humus (organic matter) content 
    bw =      ds.at[s, "Bound_water"]                # Bound water, after experiment [m3/m3]  
    CEC =     ds.at[s, "CEC"]                        # Cation Exchange Capacity [meq/100g]

    pdn =     ppm.schjonnpd(cc, org)                 # partile density of layer(s) (g/cm³)

    ################################  Fitting models evaluation ######################################
   
    etot =    np.sum((dt.loc[:, sample+"_p"]-np.mean(dt.loc[:, sample+"_p"]))**2)   # Total sum of squares
    Roth90_RMSE, LindeP_RMSE, EndresRedman_RMSE, FengSen_RMSE, Wund13_RMSE, Sen81_RMSE, Dobson_RMSE = [], [], [], [], [], [], [] # Definition of lists

    
    """ Roth's model evaluation """
    for i in range(len(alpha_)):
        
        Roth90_RMSE.append(np.sqrt(np.mean((ppm.roth( dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], alpha_[i])-dt.loc[:, sample+"_p"])**2)))    

    Roth90_RMSE_min = min(Roth90_RMSE)
    alpha_fitted = alpha_[Roth90_RMSE.index(Roth90_RMSE_min)]
    
    Roth90_fitted = ppm.roth( dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], alpha_fitted)
    RMSE.at["Roth (Eq. 23)", "RMSE_"+sample] = Roth90_RMSE_min
    R2.at["Roth (Eq. 23)", "R2_"+sample] =   1 - (np.sum((Roth90_fitted-dt.loc[:, sample+"_p"])**2)/etot)
    ds.at[s, "Alpha"] = alpha_fitted
    print("The optimum Alpha is:", alpha_fitted)
    
    
    """ Linde's model"""
    for i in range(len(m_)):
        LindeP_RMSE.append(np.sqrt(np.mean((ppm.linde(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], m_[i], m_[i])-dt.loc[:, sample+"_p"])**2)))

    m_fitted = m_[LindeP_RMSE.index(min(LindeP_RMSE))]
    
    LindeP_fitted = ppm.linde(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], m_fitted, m_fitted)
    LindeP = ppm.linde(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], 1.5, 2)
    RMSE.at["Linde (Eq. 28)", "RMSE_"+sample] = min(LindeP_RMSE)
    R2.at["Linde (Eq. 28)", "R2_"+sample]   = 1 - (np.sum((LindeP_fitted-dt.loc[:, sample+"_p"])**2)/etot)
    ds.at[s, "m_linde"] = m_fitted
    print("The optimum m is:", m_fitted)
    
    
    """ Dobson's model"""
    for i in range(len(bwp_)):
        
        Dobson_RMSE.append(np.sqrt(np.mean((ppm.dobson(dt.loc[:, sample+"_w"], bw, bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], bwp_[i]) -dt.loc[:, sample+"_p"])**2)))

    bwp_fitted = bwp_[Dobson_RMSE.index(min(Dobson_RMSE))]
    
    Dobson_fitted = ppm.dobson(dt.loc[:, sample+"_w"], bw, bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], bwp_fitted)
    RMSE.at["Dobson (Eq. 32)", "RMSE_"+sample] = min(Dobson_RMSE)
    R2.at["Dobson (Eq. 32)", "R2_"+sample] =   1 - (np.sum((Dobson_fitted-dt.loc[:, sample+"_p"])**2)/etot)
    ds.at[s, "bwp_dobson"] = bwp_fitted
    print("The optimum bwp is:", bwp_fitted)
    
    
    """ Embedding schemes evaluation"""
    for i in range(len(L_)):
        
        Sen81_RMSE.append(       np.sqrt(np.mean((ppm.sen(                dt.loc[:, sample+"_w"], bd, pdn, ap, ps,                                            dt.loc[:, sample+"_wpc"], L_[i] )-dt.loc[:, sample+"_p"])**2)))        
        FengSen_RMSE.append(     np.sqrt(np.mean((ppm.feng_sen(           dt.loc[:, sample+"_w"], bd, pdn, ap, ps,                                            dt.loc[:, sample+"_wpc"], L_[i] )-dt.loc[:, sample+"_p"])**2)))
        EndresRedman_RMSE.append(np.sqrt(np.mean((ppm.endres_redman(      dt.loc[:, sample+"_w"], bd, pdn, ap, ps,                                            dt.loc[:, sample+"_wpc"], L_[i] )-dt.loc[:, sample+"_p"])**2)))
        Wund13_RMSE.append(      np.sqrt(np.mean((ppm.wunderlich(         dt.loc[:, sample+"_w"], dt.loc[:, sample+"_p"].min(), dt.loc[:, sample+"_w"].min(), dt.loc[:, sample+"_wpc"], Lw_[i])-dt.loc[:, sample+"_p"])**2)))
    
    RMSE.at["Sen (Eq. 33)", "RMSE_"+sample]        = min(Sen81_RMSE)  
    RMSE.at["Feng&Sen (Eq. 36)", "RMSE_"+sample]      = min(FengSen_RMSE)    
    RMSE.at["Endr&Redm (Eq. 39)", "RMSE_"+sample] = min(EndresRedman_RMSE)
    RMSE.at["Wunderlich (Eq. 43)", "RMSE_"+sample]       = min(Wund13_RMSE)
    
    L_fitted_Sen81        = L_[Sen81_RMSE.index(min(Sen81_RMSE))]
    L_fitted_FengSen      = L_[FengSen_RMSE.index(min(FengSen_RMSE))]
    L_fitted_EndresRedman = L_[EndresRedman_RMSE.index(min(EndresRedman_RMSE))]
    L_fitted_Wund13       = Lw_[Wund13_RMSE.index(min(Wund13_RMSE))]

    Sen81_fitted = ppm.sen(           dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], L_fitted_Sen81)
    FengSen_fitted = ppm.feng_sen(       dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], L_fitted_FengSen)
    EndresRedman_fitted = ppm.endres_redman(  dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], L_fitted_EndresRedman)
    Wund13_fitted = ppm.wunderlich(dt.loc[:, sample+"_w"], dt.loc[:, sample+"_p"].min(), dt.loc[:, sample+"_w"].min(), dt.loc[:, sample+"_wpc"], L_fitted_Wund13)

    R2.at["Sen (Eq. 33)", "R2_"+sample]         = 1 - (np.sum((Sen81_fitted       -dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Feng&Sen (Eq. 36)", "R2_"+sample]    = 1 - (np.sum((FengSen_fitted     -dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Endr&Redm (Eq. 39)", "R2_"+sample]   = 1 - (np.sum((EndresRedman_fitted-dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Wunderlich (Eq. 43)", "R2_"+sample]  = 1 - (np.sum((Wund13_fitted      -dt.loc[:, sample+"_p"])**2)/etot)                

    ds.at[s, "L_Sen81"] = L_fitted_Sen81    
    ds.at[s, "L_FengSen"] = L_fitted_FengSen
    ds.at[s, "L_EndresRedman"] = L_fitted_EndresRedman
    ds.at[s, "L_Wund13"] = L_fitted_Wund13
    
    
    ####################################### Non-fitting models evaluation ################################
    
    length = pd.notna(dt.loc[:,sample+"_w"]).sum()
    
    crim_pred, Peplinski95_pred, Malicki_pred= [], [], [], 
    FengSen_Mod_pred, Roth90_Mod_pred, LindeP_Mod_pred, Sen81_Mod_pred, Dobson_mod_pred, EndresRedman_mod_pred, Wund13_mod_pred = [], [], [], [], [], [], []
    
    EndresRedman = ppm.endres_redman(  dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], 0.3)
    FengSen = ppm.feng_sen(       dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"],  0.3)
    Wund13 = ppm.wunderlich(dt.loc[:, sample+"_w"], 4, 0.05, dt.loc[:, sample+"_wpc"],  0.01)
    Sen81 = ppm.sen(           dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"],  0.3)
    
    R2.at["Feng&Sen (Eq. 37)", "R2_"+sample] =      1 - (np.sum((FengSen-dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Linde (Eq. 29)", "R2_"+sample] =       1 - (np.sum((LindeP - dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Sen (Eq. 34)", "R2_"+sample] =        1 - (np.sum((Sen81 - dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Endr&Redm (Eq. 40)", "R2_"+sample] = 1 - (np.sum((EndresRedman-dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Wunderlich (Eq. 44)", "R2_"+sample] =       1 - (np.sum((Wund13-dt.loc[:, sample+"_p"])**2)/etot)
    
    RMSE.at["Feng&Sen (Eq. 37)", "RMSE_"+sample] =      np.sqrt(np.mean((FengSen - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Linde (Eq. 29)", "RMSE_"+sample] =       np.sqrt(np.mean((LindeP - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Sen (Eq. 34)", "RMSE_"+sample] =        np.sqrt(np.mean((Sen81 - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Endr&Redm (Eq. 40)", "RMSE_"+sample] = np.sqrt(np.mean((EndresRedman - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Wunderlich (Eq. 44)", "RMSE_"+sample] =       np.sqrt(np.mean((Wund13 - dt.loc[:, sample+"_p"])**2))

    Hydra = ppm.hydraprobe(      dt.loc[:, sample+"_w"]  )
    Topp =  [ppm.topp(dt.loc[i, sample+"_w"]) for i in range(length)]
    Jac_SchjB =  [ppm.jacandschjB( dt.loc[i, sample+"_w"], bd, cc, org  )  for i in range(length)]
    Nadler = [ppm.nadler(  dt.loc[i, sample+"_w"]  ) for i in range(length)]
    Logsdon = [ppm.logsdonperm(  dt.loc[i, sample+"_w"] ) for i in range(length)]
    Steelman = [ppm.steelman(     dt.loc[i, sample+"_w"]  ) for i in range(length)]
    Jac_SchjA = [ppm.jacandschjA(     dt.loc[i, sample+"_w"] ) for i in range(length)]
    
    RMSE.at["HydraProbe (Eq. 20)", "RMSE_"+sample]  =          np.sqrt(np.mean((Hydra - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Topp (Eq. 15)",  "RMSE_"+sample] =           np.sqrt(np.mean(( Topp - dt.loc[:length-1, sample+"_p"])**2))
    RMSE.at["Jac&SchjB (Eq. 18)", "RMSE_"+sample] = np.sqrt(np.mean(( Jac_SchjB - dt.loc[:length-1, sample+"_p"])**2))
    RMSE.at["Nadler (Eq. 16)", "RMSE_"+sample] =          np.sqrt(np.mean(( Nadler - dt.loc[:length-1, sample+"_p"])**2))
    RMSE.at["Logsdon (Eq. 21)", "RMSE_"+sample] =     np.sqrt(np.mean(( Logsdon - dt.loc[:length-1, sample+"_p"])**2))
    RMSE.at["Steel&Endr (Eq. 22)", "RMSE_"+sample] =        np.sqrt(np.mean(( Steelman - dt.loc[:length-1, sample+"_p"])**2))
    RMSE.at["Jac&SchjA (Eq. 17)", "RMSE_"+sample] = np.sqrt(np.mean(( Jac_SchjA - dt.loc[:length-1, sample+"_p"])**2))
    
    R2.at["HydraProbe (Eq. 20)", "R2_"+sample]  =          1 - (np.sum((Hydra-dt.loc[:, sample+"_p"])**2)/etot)                                               
    R2.at["Topp (Eq. 15)",  "R2_"+sample] =           1 - (np.sum((Topp-dt.loc[:length-1, sample+"_p"])**2)/etot)
    R2.at["Jac&SchjB (Eq. 18)", "R2_"+sample] = 1 - (np.sum((Jac_SchjB-dt.loc[:length-1, sample+"_p"])**2)/etot)
    R2.at["Nadler (Eq. 16)", "R2_"+sample] =          1 - (np.sum((Nadler-dt.loc[:length-1, sample+"_p"])**2)/etot)
    R2.at["Logsdon (Eq. 21)", "R2_"+sample] =     1 - (np.sum((Logsdon-dt.loc[:length-1, sample+"_p"])**2)/etot)
    R2.at["Steel&Endr (Eq. 22)", "R2_"+sample] =        1 - (np.sum((Steelman-dt.loc[:length-1, sample+"_p"])**2)/etot)
    R2.at["Jac&SchjA (Eq. 17)", "R2_"+sample] = 1 - (np.sum((Jac_SchjA-dt.loc[:length-1, sample+"_p"])**2)/etot)

    crim_pred = ppm.crim(            dt.loc[:, sample+"_w"], bd, pdn,            ap, ps, dt.loc[:, sample+"_wpc"])
    dt['crim'+sample+'_p'] = crim_pred
    Peplinski95_pred = ppm.peplinski(       dt.loc[:, sample+"_w"], bd, pdn, cc, sc, ps, dt.loc[:, sample+"_wpc"], p_inf)
    Malicki_pred = ppm.malicki(         dt.loc[:, sample+"_w"], bd                                )
    
    RMSE.at["CRIM (Eq. 24)", "RMSE_"+sample] =            np.sqrt(np.mean((crim_pred - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Peplinski (Eq. 27)", "RMSE_"+sample] =     np.sqrt(np.mean((Peplinski95_pred - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Malicki (Eq. 19)", "RMSE_"+sample] =         np.sqrt(np.mean((Malicki_pred - dt.loc[:, sample+"_p"])**2))
    
    R2.at["CRIM (Eq. 24)", "R2_"+sample] =            1 - (np.sum((crim_pred-dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Peplinski (Eq. 27)", "R2_"+sample] =     1 - (np.sum((Peplinski95_pred-dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Malicki (Eq. 19)", "R2_"+sample] =         1 - (np.sum((Malicki_pred-dt.loc[:, sample+"_p"])**2)/etot)

    Roth90_mv =        ppm.roth_mv(dt.loc[:, sample+"_w"],      bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], CEC)
    dt['Roth_mv'+sample+'_p'] = Roth90_mv 
    LindeP_mv =        ppm.linde_mv(dt.loc[:, sample+"_w"],        bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], CEC)
    Sen81_mv =         ppm.sen_mv(dt.loc[:, sample+"_w"],         bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], CEC)
    endres_redman_mv = ppm.endres_redman_mv(dt.loc[:, sample+"_w"], bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], CEC)
    feng_sen85_mv =    ppm.feng_sen_mv(dt.loc[:, sample+"_w"],    bd, pdn, ap, ps, dt.loc[:, sample+"_wpc"], CEC)
    Wunderlich_perm_mv =     ppm.wunderlich_mv(dt.loc[:, sample+"_w"], 4, 0.05,             dt.loc[:, sample+"_wpc"], CEC)
    
    RMSE.at["Feng&Sen (Eq. 38)",   "RMSE_"+sample] =  np.sqrt(np.mean((feng_sen85_mv - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Roth (Eq. 26)",       "RMSE_"+sample] =  np.sqrt(np.mean((Roth90_mv - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Linde (Eq. 31)",      "RMSE_"+sample] =  np.sqrt(np.mean((LindeP_mv - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Sen (Eq. 35)",        "RMSE_"+sample] =  np.sqrt(np.mean((Sen81_mv - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Endr&Redm (Eq. 42)",  "RMSE_"+sample] =  np.sqrt(np.mean((endres_redman_mv - dt.loc[:, sample+"_p"])**2))
    RMSE.at["Wunderlich (Eq. 45)", "RMSE_"+sample] =  np.sqrt(np.mean((Wunderlich_perm_mv - dt.loc[:, sample+"_p"])**2))

    R2.at["Feng&Sen (Eq. 38)",   "R2_"+sample] =   1 - (np.sum((feng_sen85_mv-dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Roth (Eq. 26)",       "R2_"+sample] =   1 - (np.sum((Roth90_mv - dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Linde (Eq. 31)",      "R2_"+sample] =   1 - (np.sum((LindeP_mv - dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Sen (Eq. 35)",        "R2_"+sample] =   1 - (np.sum((Sen81_mv  - dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Endr&Redm (Eq. 42)",  "R2_"+sample] =   1 - (np.sum((endres_redman_mv-dt.loc[:, sample+"_p"])**2)/etot)
    R2.at["Wunderlich (Eq. 45)", "R2_"+sample] =   1 - (np.sum((Wunderlich_perm_mv-dt.loc[:, sample+"_p"])**2)/etot)
    
    s += 1
```

```python
'''                    Fieldwork results                 '''

# Defining a dataframe to save evaluation errors
Fits = pd.DataFrame(index=[ "crim" ,"Linde_fixed", "Endr_Redm_fixed",
                           "Roth_mv", "Linde_mv", "Endr_Redm_mv"],
                    columns = ["RMSE", "R2"])

# Evaluating the models

df["partdens"] = ppm.schjonnpd(df.Clay, df.Humus)
df['WatpermT'] = ppm.malmberg_maryott(df.field_temp)

df["crim"] = ppm.crim(df.field_water/100, df.Bulk_density, df.partdens, 1.4, ps, df.WatpermT)
df["roth_mv"] = ppm.roth_mv(df.field_water/100, df.Bulk_density, df.partdens, 1.4, ps, df.WatpermT, df.CEC_meq100g)

df["linde"] = ppm.linde(df.field_water/100, df.Bulk_density, df.partdens, 1.4, ps, df.WatpermT, 1.5, 2)
df["linde_mv"] = ppm.linde_mv(df.field_water/100, df.Bulk_density, df.partdens, 1.4, ps, df.WatpermT, df.CEC_meq100g)

df["endresredman_mv"]   =  ppm.endres_redman_mv(df.field_water/100, df.Bulk_density, df.partdens, 1.4, ps, df.WatpermT, df.CEC_meq100g)
df["endresredman"]      =  ppm.endres_redman(df.field_water/100, df.Bulk_density, df.partdens, 1.4, ps, df.WatpermT, 0.3)
    
# Error evaluation
Fits.at["crim", "R2"] = r2_score(df.crim, df.field_realperm)
Fits.at["Roth_mv", "R2"] = r2_score(df.roth_mv, df.field_realperm)

Fits.at["Linde_mv", "R2"] = r2_score(df.linde_mv, df.field_realperm)
Fits.at["Linde_fixed", "R2"] = r2_score(df.linde, df.field_realperm)

Fits.at["Endr_Redm_mv", "R2"] = r2_score(df.endresredman_mv, df.field_realperm)
Fits.at["Endr_Redm_fixed", "R2"] = r2_score(df.endresredman, df.field_realperm)

Fits.at["crim", "RMSE"] = math.sqrt(mse(df.field_realperm, df.crim))
Fits.at["Roth_mv", "RMSE"] = math.sqrt(mse(df.field_realperm, df.roth_mv))

Fits.at["Linde_mv", "RMSE"] = math.sqrt(mse(df.field_realperm, df.linde_mv))
Fits.at["Linde_fixed", "RMSE"] = math.sqrt(mse(df.field_realperm, df.linde))

Fits.at["Endr_Redm_fixed", "RMSE"] = math.sqrt(mse(df.field_realperm, df.endresredman))
Fits.at["Endr_Redm_mv", "RMSE"] = math.sqrt(mse(df.field_realperm, df.endresredman_mv))
```
