```python
ec_ = np.arange(0   , 60  , 60/len(dt.P17_w))
print(ec_)
"""  Calculating conductivity by temperature correction  """

ds["hul586_cond"] =    100000/ds.hul586_res
ds["Eref44_cond"] =    100000/ds.Eref44_res
ds["D348_cond"] =      100000/ds.D348_res
ds["Dren8_cond"] =     100000/ds.Dren8_res
ds["P17_cond"] =       100000/ds.P17_res
ds["EH23_cond"] =      100000/ds.EH23_res
ds["ValtheA11_cond"] = 100000/ds.ValtheA11_res
ds["ValtheN5_cond"] =  100000/ds.ValtheN5_res
ds["Aref44_cond"] =    100000/ds.Aref44_res
#ds["EH26_cond"] =    100000/ds.EH26_res

ds["hul586_condt"] =    ptf.arps53_cond25(ds.hul586_t, ds.hul586_cond)
ds["Eref44_condt"] =    ptf.arps53_cond25(ds.Eref44_t, ds.Eref44_cond)
ds["D348_condt"] =      ptf.arps53_cond25(ds.D348_t, ds.D348_cond)
ds["Dren8_condt"] =     ptf.arps53_cond25(ds.Dren8_t, ds.Dren8_cond)
ds["P17_condt"] =       ptf.arps53_cond25(ds.P17_t, ds.P17_cond)
ds["EH23_condt"] =      ptf.arps53_cond25(ds.EH23_t, ds.EH23_cond)
ds["ValtheA11_condt"] = ptf.arps53_cond25(ds.ValtheA11_t, ds.ValtheA11_cond)
ds["ValtheN5_condt"] =  ptf.arps53_cond25(ds.ValtheN5_cond, ds.ValtheN5_cond)
ds["Aref44_condt"] =    ptf.arps53_cond25(ds.Aref44_t, ds.Aref44_cond)
#ds["EH26_condt"] =    ptf.arps53_cond25(ds.EH26_t, ds.EH26_cond)

perm = list(range(1, 32))
ojo_all =      ptf.ojo_all(    np.array(perm))*100
ojoMB1 =       ptf.ojo_MB1(    np.array(perm))*100
ojoMB4 =       ptf.ojo_MB4(    np.array(perm))*100
ojoMB7 =       ptf.ojo_MB7(    np.array(perm))*100
ojoMB9 =       ptf.ojo_MB9(    np.array(perm))*100

offset = 6
wp = 80

"""  Calculating EC from Imaginary permittivity using Logsdon (2000) equation  """
dt["EH26_ech"]       = ptf.logsdon(fc, dt.EH26_p,       dt.EH26_i,       np)
dt["A44_ech"]       = ptf.logsdon(fc, dt.A44_p,       dt.A44_i,       np)
dt["valthe5_ech"]   = ptf.logsdon(fc, dt.valthe5_p,   dt.valthe5_i,   np)
dt["EH23_ech"]      = ptf.logsdon(fc, dt.EH23_p,      dt.EH23_i,      np)
dt["P17_ech"]       = ptf.logsdon(fc, dt.P17_p,       dt.P17_i,       np)
dt["Dren8_ech"]     = ptf.logsdon(fc, dt.Dren8_p, dt.Dren8_i, np)
dt["E44_ech"]       = ptf.logsdon(fc, dt.E44_p,       dt.E44_i,       np)
dt["D348_ech"]      = ptf.logsdon(fc, dt.D348_p,      dt.D348_i,      np)
dt["hul586_ech"]    = ptf.logsdon(fc, dt.hul586_p,    dt.hul586_i,    np)
dt["valthe11_ech"]  = ptf.logsdon(fc, dt.valthe11_p,  dt.valthe11_i,  np)

#dt["EH26_ech"]       = ptf.arps53_cond25(dt.EH26_t,       ptf.logsdon(fc, dt.EH26_p,       dt.EH26_i,       np))
#dt["A44_ech"]       = ptf.arps53_cond25(dt.A44_t,       ptf.logsdon(fc, dt.A44_p,       dt.A44_i,       np))
#dt["valthe5_ech"]   = ptf.arps53_cond25(dt.valthe5_t,   ptf.logsdon(fc, dt.valthe5_p,   dt.valthe5_i,   np))
#dt["EH23_ech"]      = ptf.arps53_cond25(dt.EH23_t,      ptf.logsdon(fc, dt.EH23_p,      dt.EH23_i,      np))
#dt["P17_ech"]       = ptf.arps53_cond25(dt.P17_t,       ptf.logsdon(fc, dt.P17_p,       dt.P17_i,       np))
#dt["Dren8_ech"] = ptf.arps53_cond25(dt.Dren8_t, ptf.logsdon(fc, dt.Dren8_p, dt.Dren8_i, np))
#dt["E44_ech"]       = ptf.arps53_cond25(dt.E44_t,       ptf.logsdon(fc, dt.E44_p,       dt.E44_i,       np))
#dt["D348_ech"]      = ptf.arps53_cond25(dt.D348_t,      ptf.logsdon(fc, dt.D348_p,      dt.D348_i,      np))
#dt["hul586_ech"]    = ptf.arps53_cond25(dt.hul586_t,    ptf.logsdon(fc, dt.hul586_p,    dt.hul586_i,    np))
#dt["valthe11_ech"]  = ptf.arps53_cond25(dt.valthe11_t,  ptf.logsdon(fc, dt.valthe11_p,  dt.valthe11_i,  np))

""" Offset calculation using Hilhorst(2000) """
regression_P17 =     linear_model.LinearRegression()
model_P17 =          regression_P17.fit(np.array([dt.P17_ech[:15]]).reshape(-1, 1)*1000, dt.P17_p[:15])
predicted_P17 =      regression_P17.predict(np.array(ec_.reshape(-1, 1)))
dt["P17_offset"] =   model_P17.intercept_

regression_E44 =     linear_model.LinearRegression()
model_E44 =          regression_E44.fit(np.array([dt.E44_ech[:13]]).reshape(-1, 1)*1000, dt.E44_p[:13])
predicted_E44 =      regression_E44.predict(np.array(ec_.reshape(-1, 1)))
dt["E44_offset"] =   model_E44.intercept_

regression_EH26 =     linear_model.LinearRegression()
model_EH26 =          regression_EH26.fit(np.array([dt.EH26_ech[:5]]).reshape(-1, 1)*1000, dt.EH26_p[:5])
predicted_EH26 =      regression_EH26.predict(np.array(ec_.reshape(-1, 1)))
dt["EH26_offset"] =   model_EH26.intercept_

regression_A44 =     linear_model.LinearRegression()
model_A44 =          regression_A44.fit(np.array([dt.A44_ech[:5]]).reshape(-1, 1)*1000, dt.A44_p[:5])
predicted_A44 =      regression_A44.predict(np.array(ec_.reshape(-1, 1)))
dt["A44_offset"] =   model_A44.intercept_

regression_valthe5 =     linear_model.LinearRegression()
model_valthe5 =          regression_valthe5.fit(np.array([dt.valthe5_ech[:14]]).reshape(-1, 1)*1000, dt.valthe5_p[:14])
predicted_valthe5 =      regression_valthe5.predict(np.array(ec_.reshape(-1, 1)))
dt["valthe5_offset"] =   model_valthe5.intercept_

regression_EH23 =     linear_model.LinearRegression()
model_EH23 =          regression_EH23.fit(np.array([dt.EH23_ech[:3]]).reshape(-1, 1)*1000, dt.EH23_p[:3])
predicted_EH23 =      regression_EH23.predict(np.array(ec_.reshape(-1, 1)))
dt["EH23_offset"] =   model_EH23.intercept_

regression_Dren8 =     linear_model.LinearRegression()
model_Dren8 =          regression_Dren8.fit(np.array([dt.Dren8_ech[:2]]).reshape(-1, 1)*1000, dt.Dren8_p[:2])
predicted_Dren8 =      regression_Dren8.predict(np.array(ec_.reshape(-1, 1)))
dt["Dren8_offset"] =   model_Dren8.intercept_

regression_D348 =     linear_model.LinearRegression()
model_D348 =          regression_D348.fit(np.array([dt.D348_ech[:2]]).reshape(-1, 1)*1000, dt.D348_p[:2])
predicted_D348 =      regression_D348.predict(np.array(np.arange(0   , 60  , 60/20).reshape(-1, 1)))
dt["D348_offset"] =   model_D348.intercept_

regression_hul586 =     linear_model.LinearRegression()
model_hul586 =          regression_hul586.fit(np.array([dt.hul586_ech[:14]]).reshape(-1, 1)*1000, dt.hul586_p[:14])
predicted_hul586 =      regression_hul586.predict(np.array(ec_.reshape(-1, 1)))
dt["hul586_offset"] =   model_hul586.intercept_

regression_valthe11 =     linear_model.LinearRegression()
model_valthe11 =          regression_valthe11.fit(np.array([dt.valthe11_ech[:14]]).reshape(-1, 1)*1000, dt.valthe11_p[:14])
predicted_valthe11 =      regression_valthe11.predict(np.array(ec_.reshape(-1, 1)))
dt["valthe11_offset"] =   model_valthe11.intercept_

""" Calculating relaxation component by substracting EC component of imaginary permittivity"""
dt["EH26_rel"] =        dt.EH26_i[:]   -     dt.EH26_ech[:]      /(2*pi*fc*e_0)
dt["A44_rel"] =        dt.A44_i[:]   -     dt.A44_ech[:]      /(2*pi*fc*e_0)
dt["valthe5_rel"] =    dt.valthe5_i[:]   - dt.valthe5_ech[:]  /(2*pi*fc*e_0)
dt["EH23_rel"] =       dt.EH23_i[:]   -    dt.EH23_ech[:]     /(2*pi*fc*e_0)
dt["P17_rel"] =        dt.P17_i[:]   -     dt.P17_ech[:]      /(2*pi*fc*e_0)
dt["Dren8_rel"] =  dt.Dren8_i[:] - dt.Dren8_ech[:]/(2*pi*fc*e_0)
dt["E44_rel"] =        dt.E44_i[:]   -     dt.E44_ech[:]      /(2*pi*fc*e_0)
dt["D348_rel"] =       dt.D348_i[:]   -    dt.D348_ech[:]     /(2*pi*fc*e_0)
dt["hul586_rel"] =     dt.hul586_i[:]   -  dt.hul586_ech[:]   /(2*pi*fc*e_0)
dt["valthe11_rel"] =   dt.valthe11_i[:]  - dt.valthe11_ech[:] /(2*pi*fc*e_0)

dt["EH26_ecw"] =        80*dt.EH26_ech[:]      /(dt.EH26_p - dt.EH26_offset)
dt["A44_ecw"] =        80*dt.A44_ech[:]      /(dt.A44_p - dt.A44_offset)
dt["valthe5_ecw"] =    80*dt.valthe5_ech[:]  /(dt.valthe5_p - dt.valthe5_offset)
dt["EH23_ecw"] =       80*dt.EH23_ech[:]     /(dt.EH23_p - dt.EH23_offset)
dt["P17_ecw"] =        80*dt.P17_ech[:]      /(dt.P17_p - dt.P17_offset)
dt["Dren8_ecw"] =  80*dt.Dren8_ech[:]/(dt.Dren8_p - dt.Dren8_offset)
dt["E44_ecw"] =        80*dt.E44_ech[:]      /(dt.E44_p - dt.E44_offset)
dt["D348_ecw"] =       80*dt.D348_ech[:]     /(dt.D348_p - dt.D348_offset)
dt["hul586_ecw"] =     80*dt.hul586_ech[:]   /(dt.hul586_p - dt.hul586_offset)
dt["valthe11_ecw"] =   80*dt.valthe11_ech[:] /(dt.valthe11_p - dt.valthe11_offset)

dt.to_excel("calibration_curves_IOP.xlsx")
dt
```
