```python
"""     Reading the data Samples_database   """
df = pd.read_excel("Samples_database_.xlsx")[:62]
df[0:20]
```

```python
glass_mass = 10.773
glass_vol =  117.8

df["Porosity1"] =      1 - ((df.soil_plastic_glass_mass1-glass_mass)/glass_vol)/2.65
df["Es_1_CRIM"] =           ptf.crim_es(    df.Real_perm1, 1.4, ((df.soil_plastic_glass_mass1-glass_mass)/glass_vol), 2.65)    
df["Porosity2"] =      1 - ((df.glass_dry_sample2-glass_mass)/glass_vol)/2.65
df["Es_2_CRIM"] =           ptf.crim_es(    df.Real_perm2, 1.4, ((df.glass_dry_sample2-glass_mass)/glass_vol), 2.65)    
df["Es_2_HBS"] =       ptf.HBS_es((df.glass_dry_sample2-glass_mass)/glass_vol, 2.65, df.Real_perm2, 1.4)
df["Es_dif"] =         df.Es_2_CRIM - df.Es_1_CRIM
df["mass_loss"] =      df.cup_wet_sample2 - df.cup_dry_sample2
df["mass_loss_prop"] = df.dry_sample0 * df.mass_loss/(df.cup_wet_sample2 - df.cup_mass)
df["bound_water"] =    df.mass_loss_prop/df.copecky_vol
df["bound_water100"] = df.bound_water*100
df["bw_plus_cec"] =    df.bound_water/df.bound_water.max() + df.CEC_meq100g/df.CEC_meq100g.max()

df["part_dens"] =      ptf.schjonnpd(df.Clay, df.Humus)

df[1:60]
```
